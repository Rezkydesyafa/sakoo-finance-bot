# PARSER-005 - Parser service output JSON dan confidence

Notebook ini mendemonstrasikan service `parse_message(text, source)` yang menyatukan intent, amount, date, type, category classifier, confidence, dan `need_confirmation` menjadi satu output final sesuai struktur PRD.

```python
from datetime import date
from pathlib import Path
import json
import sys

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
BACKEND = ROOT / "backend"
sys.path.insert(0, str(BACKEND))

from app.modules.parser.service import parse_message
```

## Contoh transaksi PRD

Confidence `>= 0.85` berarti output cukup yakin untuk auto-save. Confidence rendah akan mengaktifkan `need_confirmation`.

```python
TODAY = date(2026, 6, 27)
examples = [
    "beli makan 20 ribu",
    "bayar bensin 30rb kemarin",
    "gaji masuk 2 juta",
    "beli buku rp20.000 tanggal 26/06/2026",
]

for text in examples:
    result = parse_message(text, source="whatsapp_text", today=TODAY)
    print(text)
    print(json.dumps(result.to_dict(), indent=2, ensure_ascii=False))
    print()
```

## Contoh command bot

Command laporan/export/riwayat/help tidak boleh salah diproses sebagai transaksi.

```python
commands = [
    "laporan bulan ini",
    "export laporan bulan ini",
    "riwayat transaksi",
    "hapus transaksi terakhir",
    "bantuan",
]

for text in commands:
    result = parse_message(text, source="whatsapp_text", today=TODAY)
    print(text)
    print(json.dumps(result.to_dict(), indent=2, ensure_ascii=False))
    print()
```

## Contoh confidence rendah

Input yang nominalnya ada tetapi kategori masih fallback akan meminta konfirmasi.

```python
ambiguous = parse_message("keluar 20 ribu", source="whatsapp_text", today=TODAY)
print(json.dumps(ambiguous.to_dict(), indent=2, ensure_ascii=False))
```
